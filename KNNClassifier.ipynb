```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
%matplotlib inline
```

```python
x,y=make_classification(n_samples=1000, n_features=3,n_classes=2, n_redundant=1,random_state=999)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x
```

```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
```

```python
classifier=KNeighborsClassifier(n_neighbors=5,algorithm='auto')
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
param_grid = {'n_neighbors': list(range(1, 20))}  # test K from 1 to 20
```

```python
knn = KNeighborsClassifier()
```

```python
grid = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')  # 5-fold CV
grid.fit(x_train, y_train)
```

```python
y_predd= grid.predict(x_test)
```

##KNN regressor problem

```python
from sklearn.datasets import make_regression
```

```python
X,Y=make_regression(n_samples=1000,n_features=2,noise=10,random_state=42)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
regress=KNeighborsRegressor(n_neighbors=6,algorithm='auto')
regress.fit(X_train,Y_train)
```

```python
Y_pred=regress.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(Y_pred,Y_test))
print(mean_absolute_error(Y_pred,Y_test))
print(mean_squared_error(Y_pred,Y_test))
```
