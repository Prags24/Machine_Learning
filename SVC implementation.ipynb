```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,Y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
Y
```

```python
a=pd.DataFrame(X)[0]
b=pd.DataFrame(X)[1]
```

```python
sns.scatterplot(x=a,y=b,hue=Y)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
svc=SVC(kernel='linear' )
```

```python
svc.fit(X_train,Y_train)
```

```python
##Prediction
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```

```python
rbf=SVC(kernel='rbf' )
```

```python
rbf.fit(X_train,Y_train)
```

```python
##Prediction
y_pred1=rbf.predict(X_test)
```

```python
print(confusion_matrix(Y_test,y_pred1))
print(classification_report(Y_test,y_pred1))
```

```python
poly=SVC(kernel='poly')
poly.fit(X_train,Y_train)
y_pred2=poly.predict(X_test)
print(confusion_matrix(Y_test,y_pred2))
print(classification_report(Y_test,y_pred2))
```

```python
sigm=SVC(kernel='sigmoid')
sigm.fit(X_train,Y_train)
y_pred3=sigm.predict(X_test)
print(confusion_matrix(Y_test,y_pred3))
print(classification_report(Y_test,y_pred3))
```

##rbf is giving the best accuracy
**Now with hyperparameter tuning***

```python

from sklearn.model_selection import GridSearchCV

##defining parameter range
param_grid={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
y_pred4=grid.predict(X_test)
print(confusion_matrix(Y_test,y_pred4))
print(classification_report(Y_test,y_pred4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#outer coordinates
x=np.linspace(-5.0,5.0,100)
y=np.sqrt(10**2 - x**2)
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

```python
#inner coordinates
x1=np.linspace(-5.0,5.0,100)
y1=np.sqrt(5**2 - x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

```python
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
x
```

```python
df1=pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y']=0
df2=pd.DataFrame(np.vstack([y1,x1]).T,columns=['X1','X2'])
df2['Y']=1
df = pd.concat([df1, df2], ignore_index=True)
df.tail()
```

```python
##Independent and dependent features
X=df.iloc[:,:2]
Y=df.Y
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

**Aplying Polynomial kernel**

```python
##we need to find components for polynomial kernel
#x1,x2,x1sqr,x2sqr,x1*x2
df['x1_square']=df['X1']**2
df['x2_square']=df['X2']**2
df['x1x2']=(df['X1']*df['X2'])
df.head()
```

```python
##Independent and dependent features
X=df[['X1','X2','x1_square','x2_square','x1x2']]
Y=df['Y']
```

```python
##Independent and dependent features
X=dff[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
Y=dff['total_bill']
```

```python

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
X_train.head()
```

```python
##Feature Encoding(Label encoding and onehot encoding)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['time']=le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
X_test['sex']=le1.transform(X_test['sex'])
X_test['smoker']=le2.transform(X_test['smoker'])
X_test['time']=le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
##Onehot encoding -----column transformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])], remainder= 'passthrough' )
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)

X_train=ct.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=ct.transform(X_test)
```

##Support vector regression

```python
from sklearn.svm import SVR
svr=SVR()
```

```python
svr.fit(X_train,Y_train)
```

```python
y_pred=svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(Y_test,y_pred))
print(mean_absolute_error(Y_test,y_pred))
      
```

```python
##Hyperparameter tuning using grid search cv
from sklearn.model_selection import GridSearchCV

##defining parameter range
param_grid={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf']}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose=3)
##fitting the model for grid search
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
grid_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(Y_test,grid_pred))
print(mean_absolute_error(Y_test,grid_pred))
```
