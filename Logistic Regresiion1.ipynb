```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
##create a dataset
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
x
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#Model Training
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred=logistic.predict(x_test)
y_pred
```

```python
logistic.predict_proba(x_test)
#whichever will have high probability that will be the outcome
```

```python
##Performance metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
```

```python
print(score)
print(cm)
print(classification_report(y_test,y_pred))
```

##Hyperparameter tuning and Cross validation

```python
model=LogisticRegression()
penaltty=['l1','l2','elasticnet']
c_values=[100, 10, 1.0, 0.1, 0.01]
solvers=['newton-cg','lbfgs','liblinear','sag','saga']

params=dict(penalty=penaltty,C=c_values,solver=solvers)
```

```python
#GRid search
from sklearn.model_selection import GridSearchCV, StratifiedKFold
CV=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=CV,n_jobs=-1)

```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

##Logistic Regression for imbalanced dataset

```python
##Generate and plot a synthetic imbalanced classification dataset
from collections import Counter
from sklearn.datasets import make_classification
```

```python
##imbalanced datasets
x,y=make_classification(n_samples=10000, n_features=2,n_clusters_per_class=1, n_redundant=0,weights=[0.99],random_state=10)
```

```python
x
```

```python
Counter(y)
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y) 
plt.show()
```

```python
#split the dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
class_weight
```

```python
#Hyperparameter tuning
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']

params=dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

##Logistic Regression with ROC curve and ROC AUC score

```python
from sklearn.metrics import roc_curve, roc_auc_score
from matplotlib import pyplot
import matplotlib.pyplot as plt
```

```python
#generate 2 class dataset
X,Y= make_classification(n_samples=1000,n_classes=2,random_state=1)
```

```python
##Split the dataset into train and test
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=1)
```

```python
#create a dummy model with default 0 as output 
dummy_model_prob=[0 for _ in range(len(Y_test))]
dummy_model_prob
```

```python
##Lets create basic logistic model
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
##Prediction based on probability
model_prob=model.predict_proba(X_test)
```

```python
model_prob
```

```python
##Lets focus on positive outcome
model_prob=model_prob[:,1]
```

```python
model_prob
```

```python
##Lets calculate the scores
dummy_model_auc=roc_auc_score(Y_test,dummy_model_prob)
model_auc=roc_auc_score(Y_test,model_prob)
print(dummy_model_auc)
print(model_auc)
```

##Roc curve is the curve between False positive rate on x axis and True positive rate on y axis

```python
##Calculate ROC curves
dummy_fpr,dummy_tpr,_=roc_curve(Y_test,dummy_model_prob)
model_fpr,model_tpr,thresholds=roc_curve(Y_test,model_prob)
```

```python
##plot the roc curve for the model
pyplot.plot(dummy_fpr,dummy_tpr,linestyle=':',label='Dummy Model')
pyplot.plot(model_fpr,model_tpr,linestyle='-',label='Logistic')
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
pyplot.legend()
pyplot.show()
```

##here we can see the area under the logistic curve is more than area under dummy model which shows that our cuve has performed well

```python
thresholds
```

```python
##plot the roc curve for the model
import numpy as np
fig=pyplot.figure(figsize=(20,60))
pyplot.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
pyplot.plot(model_fpr,model_tpr,linestyle='-',label='Logistic')
ax=fig.add_subplot(111)
for xyz in zip(model_fpr,model_tpr,thresholds):
    ax.annotate('%s' % np.round(xyz[2],2), xy=(xyz[0],xyz[1]))
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
pyplot.legend()
pyplot.show()
```
